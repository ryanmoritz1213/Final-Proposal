# Fourier Optics and Uncertainty Principle 

## Description


FourierBound is a Python toolset built to map the Heisenberg Uncertainty Principle onto a measurable optical system. In this framework, we treat the physical width of a slit as the spatial constraint (Δx) and use a Fast Fourier Transform (FFT) to compute the resulting diffraction pattern, which represents the momentum spread (Δp). The core of the project is a wave propagation engine that simulates how light behaves as it passes through an aperture, allowing for a direct numerical test of the ΔxΔp≥λ/4π limit.

Implementation and Application
This code makes it very convenient to do these kind of calculate and make these simujlations. By inputting specific hardware parameters like the laser's wavelength and the camera’s pixel it can compare theoretical diffraction against real world measurements. This approach helps identify how much experimental factors like sensor quantization and beam misalignment actually skew the uncertainty product.
